```python
# sierpinski triangle 
import torch
import matplotlib.pyplot as plt

# Device 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Image size
size = 512
x, y = torch.meshgrid(
    torch.arange(size, device=device),
    torch.arange(size, device=device),
    indexing="ij"
)

# XOR rule creates the Sierpinski triangle
pattern = (x & y) == 0  

# Plot
plt.figure(figsize=(6,6))
plt.imshow(pattern.cpu(), cmap="binary")
plt.axis("off")
plt.title("Sierpinski Triangle (PyTorch)")
plt.show()
```

I used PyTorch tensors on GPU. The Sierpinski pattern comes from applying a bitwise AND (x & y) across a full tensor grid in parallel, which is much faster than pixel-by-pixel loops. 
It’s generated by a simple repeating rule that removes the center of a equilateral triangle at each step, creating a self-similar pattern.
